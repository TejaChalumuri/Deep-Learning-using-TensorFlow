```python
#as we can see from the above images
# we have outliers in the data
# So before going forward to EDA we will first remove the outliers

# creating function to remove outliers
def treat_outlier(col):
    sorted(col)
    Q1,Q3=np.percentile(col,[25,75])
    IQR=Q3-Q1
    lower_range= Q1-(1.5 * IQR)
    upper_range= Q3+(1.5 * IQR)
    return lower_range, upper_range
```

```python
# Removing outliers by applicy the fuction
for feature in df_mar[['fare_amount', 'extra', 'mta_tax', 'total_amount']]:
    lr,ur=treat_outlier(df_mar[feature])
    df_mar[feature]=np.where(df_mar[feature]>ur,ur,df_mar[feature])
    df_mar[feature]=np.where(df_mar[feature]<lr,lr,df_mar[feature])
```

```python
# Removing outliers by applicy the fuction
for feature in df_may[['fare_amount', 'extra', 'mta_tax', 'total_amount']]:
    lr,ur=treat_outlier(df_may[feature])
    df_may[feature]=np.where(df_may[feature]>ur,ur,df_may[feature])
    df_may[feature]=np.where(df_may[feature]<lr,lr,df_may[feature])
```

```python
# Removing outliers by applicy the fuction
for feature in df_jan[['fare_amount', 'extra', 'mta_tax', 'total_amount']]:
    lr,ur=treat_outlier(df_jan[feature])
    df_jan[feature]=np.where(df_jan[feature]>ur,ur,df_jan[feature])
    df_jan[feature]=np.where(df_jan[feature]<lr,lr,df_jan[feature])
```
